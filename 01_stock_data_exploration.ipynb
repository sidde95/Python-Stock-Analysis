# Different types of Plot in Stock Analysis

- **Line plot (price over time)**: close, moving averages, indicator lines.
- **Candlestick / OHLC (price action)**: open-high-low-close per day (Plotly is especially good here).
- **Area plot**: like line, but filled (good for cumulative returns, equity curve).
- **Bar chart (volume)**: daily volume bars under the price chart.
- **Histogram**: distribution of daily returns, volume, spreads.
- **KDE / density plot**: smoother distribution of returns (seaborn excels).
- **Box plot / violin plot**: compare distributions (returns by month, by ticker).
- **Scatter plot**: return vs volume, return vs volatility, feature relationships.
- **Heatmap**: correlation matrix of returns across tickers/features.
- **Pairplot (seaborn)**: quick multivariate scatter/hist grid.

```python
# Data Flattening
data.reset_index(inplace = True)

data.head()
```

```python
data.index
```

```python
# Flatten MultiIndex columns
data.columns = ['_'.join(col).strip('_') for col in data.columns]
data.head()
```

```python
# 1. Line Plot (Close Price with Moving Averages)
# Calculate moving averages
data['MA_20'] = data['Close_AAPL'].rolling(window=20).mean()
data['MA_50'] = data['Close_AAPL'].rolling(window=50).mean()

fig = go.Figure()

# Add Close Price line
fig.add_trace(go.Scatter(
    x=data['Date'], 
    y=data['Close_AAPL'],
    mode='lines',
    name='Close Price',
    line=dict(color='blue', width=2)
))

# Add 20-day Moving Average
fig.add_trace(go.Scatter(
    x=data['Date'],
    y=data['MA_20'],
    mode='lines',
    name='20-Day MA',
    line=dict(color='orange', width=2, dash='dash')
))

# Add 50-day Moving Average
fig.add_trace(go.Scatter(
    x=data['Date'],
    y=data['MA_50'],
    mode='lines',
    name='50-Day MA',
    line=dict(color='red', width=2, dash='dash')
))

fig.update_layout(
    title='AAPL Close Price with Moving Averages',
    xaxis_title='Date',
    yaxis_title='Price ($)',
    hovermode='x unified',
    height=600,
    template='plotly_white'
)

fig.show()
```

```python
# 2. Candlestick Chart
fig = go.Figure(data=[go.Candlestick(
    x=data['Date'],
    open=data['Open_AAPL'],
    high=data['High_AAPL'],
    low=data['Low_AAPL'],
    close=data['Close_AAPL']
)])

fig.update_layout(title='AAPL(Apple) Candlestick Chart', 
                  xaxis_title='Date', 
                  yaxis_title='Price ($)',
                  height=600)
fig.show()
```

```python
# 3. Area Chart (Close Price)
fig = px.area(data, 
              x='Date', 
              y='Close_AAPL',
              title='AAPL Close Price Over Time',
              labels={'Close_AAPL': 'Close Price ($)',
                      'Date': 'Date'})

fig.update_layout(
    hovermode='x unified',
    height=600,
    yaxis_title='Price ($)',
    xaxis_title='Date'
)

fig.show()
```

```python
# 4. Bar Chart (Volume) + Line Chart (Price)
from plotly.subplots import make_subplots

fig = make_subplots(rows=2, cols=1, 
                    subplot_titles=("AAPL Close Price", "Daily Trading Volume"),
                    shared_xaxes=True,
                    row_heights=[0.7, 0.3],
                    vertical_spacing=0.15)

# Add price line chart on top
fig.add_trace(
    go.Scatter(x=data['Date'], y=data['Close_AAPL'], 
               mode='lines', name='Close Price', 
               line=dict(color='blue', width=2)),
    row=1, col=1
)

# Add volume bars on bottom
fig.add_trace(
    go.Bar(x=data['Date'], y=data['Volume_AAPL'], 
           name='Volume', marker=dict(color='lightblue')),
    row=2, col=1
)

fig.update_yaxes(title_text="Price ($)", row=1, col=1)
fig.update_yaxes(title_text="Volume", row=2, col=1)
fig.update_xaxes(title_text="Date", row=2, col=1)

fig.update_layout(title_text="AAPL Stock Price and Trading Volume", 
                  height=800, hovermode="x unified")
fig.show()
```

```python
# 5. Histogram Plot (Daily Returns Distribution)
# Calculate daily returns
data['Daily_Return'] = data['Close_AAPL'].pct_change() * 100

fig = px.histogram(data, x='Daily_Return', nbins=30, 
                   title='Distribution of AAPL Daily Returns',
                   labels={'Daily_Return': 'Daily Return (%)'})

fig.update_layout(
    xaxis_title='Daily Return (%)',
    yaxis_title='Frequency',
    height=600
)

fig.show()
```

```python
# 6. KDE Plot
plt.figure(figsize=(12, 6))

data['Price_Diff'] = data['Open_AAPL'] - data['Close_AAPL']

sns.kdeplot(data=data, x='Price_Diff', fill=True, alpha=0.6, color='steelblue')

plt.title('Distribution of AAPL Open vs Close Price Difference')  
plt.xlabel('Price Difference ($)') 
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
# Box Plot
# 7. Violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(data=data[['Close_AAPL', 'Open_AAPL']], palette="Set2")
plt.title('Violin Plot of AAPL Open and Close Prices')
plt.ylabel('Price ($)')
plt.tight_layout()
plt.show()
```

```python
# 8. Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Open_AAPL', y='Close_AAPL', 
                color='blue', alpha=0.6, s=100)
plt.title('Scatter Plot of AAPL Open vs Close Prices')
plt.xlabel('Open Price ($)')
plt.ylabel('Close Price ($)')
plt.tight_layout()
plt.show()
```

```python
# 9. Heatmap: Correlation Matrix
plt.figure(figsize=(10, 8))
correlation_matrix = data[['Open_AAPL', 'High_AAPL', 'Low_AAPL', 'Close_AAPL', 'Volume_AAPL']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', center=0, fmt='.2f', 
            square=True, linewidths=1, cbar_kws={'label': 'Correlation'})
plt.title('Correlation Matrix of AAPL Stock Features')
plt.tight_layout()
plt.show()
```

```python
# 10. Pairplot (Seaborn): Multivariate Scatter/Hist Grid
pairplot_data = data[['Open_AAPL', 'High_AAPL', 'Low_AAPL', 'Close_AAPL', 'Volume_AAPL']].copy()
pairplot_data.columns = ['Open', 'High', 'Low', 'Close', 'Volume']

sns.pairplot(pairplot_data, diag_kind='hist', plot_kws={'alpha': 0.6, 's': 10})
plt.suptitle('AAPL Stock Features - Pairplot (Scatter Matrix)', y=1.001)
plt.tight_layout()
plt.show()
```
